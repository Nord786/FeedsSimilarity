```python
import sys
import json
import justext
import gensim
import itertools

import logging
import operator
import re

import nltk
from nltk.collocations import TrigramCollocationFinder
from nltk.metrics import BigramAssocMeasures, TrigramAssocMeasures

import gensim
from collections import Counter



reload(logging)
FORMAT = '%(asctime)-15s %(message)s'
logging.basicConfig(format=FORMAT, level=logging.INFO)

def get_text(html_text):
    res = ''
    paragraphs = justext.justext(html_text, justext.get_stoplist('Russian'))
    for paragraph in paragraphs:
        if not paragraph.is_boilerplate:
            res += paragraph.text
    return res

def tokenize(text):
    return gensim.utils.tokenize(text,  lower=True)

def split_words(text, stopwords=[]):
    """
    Break text into a list of single words. Ignore any token that falls into
    the `stopwords` set.

    """
    return [word
            for word in tokenize(text)
            if word not in stopwords and len(word) > 3]

def tokenize_new(text, stopwords, bigrams_pat):
    """
    Break text (string) into a list of Unicode tokens.
    
    The resulting tokens can be longer phrases (collocations) too,
    e.g. `new_york`, `real_estate` etc.

    """
    text = u' '.join(split_words(text, stopwords))    
    text = re.sub(bigrams, lambda match: match.group(0).replace(u' ', u'_'), text)
    return text.split()

logging.info("OLOLO")
```

```python
def add_text(file_name, LOG_EVERY=100):
    """
    Функция извлекает из контента текст и записывает в поле "Text"
    и сохраняет в файл с суффиксом .with_text
    """
    fin_dump = open(file_name)
    fout_dump = open(file_name + ".with_text", 'w')
    
    for i, line in enumerate(fin_dump):
        if LOG_EVERY > 0 and i % LOG_EVERY == 0:             
            logging.info("Complete documents %4d" % i)
        url, data = line.strip().split('\t', 1)
        data = json.loads(data)
        text = get_text(data['Content'])
        data['Text'] = text
        fout_dump.write("%s\t%s\n" % (url, json.dumps(data) ) )
    fin_dump.close()
    fout_dump.close()
    
#add_text('dump_100.json.sampling')
```

```python
import nltk
from nltk.collocations import TrigramCollocationFinder
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures, TrigramAssocMeasures

def best_ngrams(words, top_n=1000, min_freq=100):
    """
    Extract `top_n` most salient collocations (bigrams and trigrams),
    from a stream of words. Ignore collocations with frequency
    lower than `min_freq`.

    This fnc uses NLTK for the collocation detection itself -- not very scalable!

    Return the detected ngrams as compiled regular expressions, for their faster
    detection later on.

    """
    tcf = TrigramCollocationFinder.from_words(words)
    tcf.apply_freq_filter(min_freq)
    trigrams = [' '.join(w) for w in tcf.nbest(TrigramAssocMeasures.chi_sq, top_n)]
    logging.info("%i trigrams found: %s..." % (len(trigrams), list(trigrams)[:20]))

    bcf = tcf.bigram_finder()
    bcf.apply_freq_filter(min_freq)
    bigrams = [' '.join(w) for w in bcf.nbest(BigramAssocMeasures.pmi, top_n)]
    logging.info("%i bigrams found: %s..." % (len(bigrams), list(bigrams)[:20]))

    pat_gram2 = re.compile('(%s)' % '|'.join(bigrams), re.UNICODE)
    pat_gram3 = re.compile('(%s)' % '|'.join(trigrams), re.UNICODE)

    return pat_gram2, pat_gram3


def best_bigram(words, top_n=200, min_freq=10):
    """
    Extract `top_n` most salient collocations (bigrams and trigrams),
    from a stream of words. Ignore collocations with frequency
    lower than `min_freq`.

    This fnc uses NLTK for the collocation detection itself -- not very scalable!

    Return the detected ngrams as compiled regular expressions, for their faster
    detection later on.

    """    
    bcf = BigramCollocationFinder.from_words(words)
    bcf.apply_freq_filter(min_freq)
    bigrams = [' '.join(w) for w in bcf.nbest(BigramAssocMeasures.pmi, top_n)]
    logging.info("%i bigrams found: %s..." % (len(bigrams), list(bigrams)[:20]))

    pat_gram2 = re.compile('(%s)' % '|'.join(bigrams), re.UNICODE)    

    return pat_gram2, bigrams

bigrams_pat, bigrams = best_bigram(words_seq, 200, 10)
print bigrams_pat
```

```python
from collections import defaultdict
EVERY=100
index = defaultdict(lambda : defaultdict(float))
res = defaultdict(list)

texts_doc2bow = []
for i, (url, feed, text) in enumerate(texts): 
    if i % EVERY == 0: logging.info(i)
    texts_doc2bow.append( id2word.doc2bow(tokenize_new(text, stopwords, bigrams_pat)) )   
```

```python
import operator
TOP_N = 50

for i, (i_url, i_feed, i_text) in enumerate(texts):
    if i % EVERY == 0: logging.info(i)
    i_bow = texts_doc2bow[i]
    for j, (j_url, j_feed, j_text) in enumerate(texts):
        if i <= j: continue
        j_bow = texts_doc2bow[j]
        messure = gensim.matutils.cossim(i_bow, j_bow)
        if messure <= 0.01: continue
        res[(i_url, i_feed)].append( (j_url, j_feed, messure) )
        res[(j_url, j_feed)].append( (i_url, i_feed, messure) )
    res[(i_url, i_feed)].sort(key=operator.itemgetter(2), reverse=True)
    res[(i_url, i_feed)] = res[(i_url, i_feed)][:TOP_N]
else:
    logging.info(i)
```
