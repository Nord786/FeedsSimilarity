```python
import sys
import json
import justext
import gensim
import itertools

import logging
import operator
import re


reload(logging)
FORMAT = '%(asctime)-15s %(message)s'
logging.basicConfig(format=FORMAT, level=logging.INFO)

def tokenize_raw(text):
    return gensim.utils.tokenize(text,  lower=True)

def split_words(text, stopwords=[]):
    """
    Break text into a list of single words. Ignore any token that falls into
    the `stopwords` set.
    """
    return [word
            for word in tokenize_raw(text)
            if word not in stopwords and len(word) > 3]

def tokenize_new(text, stopwords, bigrams_pat):
    """
    Break text (string) into a list of Unicode tokens.
    
    The resulting tokens can be longer phrases (collocations) too,
    e.g. `new_york`, `real_estate` etc.

    """
    text = u' '.join(split_words(text, stopwords))    
    text = re.sub(bigrams_pat, lambda match: match.group(0).replace(u' ', u'_'), text)
    return text.split()

logging.info("OLOLO")
```

Загружаем заготовленные stopwords и регулярки для bigram

```python
import cPickle as pickle
stopwords = pickle.load(open('./data/stopwords.pkl'))
bigrams_pat = pickle.load(open('./data/bigrams_pat.pkl'))
```

Загружаем тексты и аггрегируем по feed

```python
from collections import defaultdict

texts_urls = {}
texts_feeds = defaultdict(list)

fin_dump = open('./data/dump_100.json.sampling.with_text')
for i, line in enumerate(fin_dump):
    if i % 1000 == 0:         
        logging.info("Complete documents %4d" % i)
        
    parts = line.strip().split('\t', 1)
    url, feed = parts[0].split(' : ')
    text = tokenize_new(json.loads(parts[1])['Text'], stopwords, bigrams_pat)
    texts_urls[ (url, feed) ] = text
    texts_feeds[ feed ] += text
else:
    logging.info("Complete documents %4d" % i)
```

Строим словарь по текстам статей (не фидов, чтобы удобнее было потом прочищать по кол-ву документов в которых встретилось)

```python
%time id2word = gensim.corpora.Dictionary(documents=texts_urls.itervalues())
print id2word
# ignore words that appear in less than 10 documents or more than 10% documents
print id2word.filter_extremes(no_below=10, no_above=0.1)
print id2word
```

Конвертируем тексты feed в bag of words для дальнейшего использования в моделях

```python
bow_feeds = dict()
for feed, text in texts_feeds.iteritems():
    bow_feeds[feed] = id2word.doc2bow(text)
```

Рассчитываем расстояние между всеми парами фидов (составляем матрицу similarity)

```python
feeds_similarity = defaultdict(list)

MIN_SIMILARITY=0.01
LOG_EVERY = 10

for i, (i_feed, i_bow_feed) in enumerate(bow_feeds.iteritems()):
    if i % LOG_EVERY == 0:
        logging.info(i)
    for j,(j_feed, j_bow_feed) in enumerate(bow_feeds.iteritems()):
        if i <= j: continue        
        sim = gensim.matutils.cossim(i_bow_feed, j_bow_feed)
        feeds_similarity[i_feed].append( (j_feed, sim) )
        feeds_similarity[j_feed].append( (i_feed, sim) )
else:
    logging.info(i)
    
for feed, similar_feeds in feeds_similarity.iteritems():
    similar_feeds.sort(key=operator.itemgetter(1), reverse=True)
```

Выводим для каждого фида 5 наиболее похожих со степенью похожести

```python
NUM_MOST_SIM = 5
for feed, similar_feeds in feeds_similarity.iteritems():
    print feed
    for most_similar_feed in similar_feeds[:NUM_MOST_SIM]:
#         print most_similar_feed
        print "\t%s\t%s" % most_similar_feed
```

## Similarity через встроенную MatrixSimilarity и различные модели

Подготовка

```python
from gensim import similarities
feeds_keys, feeds_corpus = zip(*bow_feeds.iteritems())
```

### BOW Model

```python
bow_sim_matrix = similarities.MatrixSimilarity(feeds_corpus)
```

```python
for feed_key, feed_bow in zip(feeds_keys, feeds_corpus):
    print feed_key
    for doc_id, sim in sorted(enumerate(bow_sim_matrix[feed_bow]), key=operator.itemgetter(1), reverse=True)[:10]:
        print "  %s\t%s" % (feeds_keys[doc_id], sim)
    print '-*-'*7
        
```

### TfIdf Model

```python
%time tfidf_model = gensim.models.TfidfModel(feeds_corpus, id2word=id2word)
```

```python
tfidf_sim_matrix = similarities.MatrixSimilarity(tfidf_model[feeds_corpus])
```

```python
def print_similarity(keys, corpus, model, sim_matrix, top_n, min_sim=0):
    for key, item in zip(keys, corpus):
        print key
        for doc_id, sim in sorted(enumerate(sim_matrix[model[item]]), key=operator.itemgetter(1), reverse=True)[:top_n]:
            if sim < min_sim: break
            sim_key = keys[doc_id]
            if key == sim_key: continue
            print "  %s\t%s" % (sim_key, sim)
        print '-*-'*7
    
print_similarity(feeds_keys, feeds_corpus, tfidf_model, tfidf_sim_matrix, 10, 0.01)
```
